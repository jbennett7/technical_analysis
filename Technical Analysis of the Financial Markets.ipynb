# Technical Analysis of the Financial Markets

The three premises on which technical analysis is based on:
1. Market action discounts everything.
2. Prices move in trends.
3. History repeats itself.

## Dow Theory

1. The averages discount everything.
2. The market has three trends.
  * __Primary trends__ are trends that last longer than a year.
  * __Secondary (or intermediate) trends__ are trends that last between three weeks to three months.
  * __Minor trends__ are trends that last less than three weeks.
3. Major trends have three phases.
  * accumulation phase
  * public particiapation phase
  * distribution phase
4. The averages must confirm each other.
  * The averages are the industrial and transport Dow averages.
5. Volume must confirm the trend.
6. A trend is assumed to be in effect until it gives definite signals that it has reversed.

## Basic Concepts of Trend

### Trend has Three Directions

#### Uptrend

As defined in Dow Theory, an uptrend is identified by three successive peaks and troughs, each higher than the last. This charts an uptrend for the Schwab Growth ETF, trading symbol 'SCHG'. The trend is illustrated with the simple moving average trend line which starts around February 6 and ends February 19.

```r
#TODO: How do you plot multiple plots with chartSeries.
#      Want to add points at the trend peaks and troughs.
symbols=c('SCHG')
getSymbols(symbols, from="2009-10-01", to="2011-10-20")
plot.data<-Cl(SCHG)
options(repr.plot.width=12, repr.plot.height=8)
chartSeries(
    plot.data["2010-02-01/2010-03-01"],
    name="Uptrends are defined by three successively higher peaks and troughs",
    theme=chartTheme('white'),
    legend=NULL,
    TA=c("addTA(SMA(plot.data,3),col='blue', on=1, legend=NULL)")
)
rm(SCHG)
```

#### Down Trend

Down trends are more pronounced, having steeper falls. Here is a downtrend for Khols, trading symbol KSS. The SMA line illustrates the trend downward.

```r
symbols=c('KSS')
getSymbols(symbols, from="2015-04-01", to="2015-11-01")
plot.data<-Cl(KSS)
options(repr.plot.width=12, repr.plot.height=8)
chartSeries(plot.data,theme=chartTheme('white'),
    TA="addTA(SMA(plot.data,40), on=1, col='blue')")
rm(KSS)
```

### Three Types of Trends

Trends are classified by the period of time that they follow. These classifications are defined in Dow theory and are:
1. __major__ - last longer than a year
2. __intermediate__ - last from about three weeks to many months (less than a year).
3. __near term__ - less than three weeks

#### Major Trend

Again using General Electric as an example, you can see the sideways trend, from about July 2004 to the low in July 2007 as a major sideways trend.

```r
symbols=c('GE')
getSymbols(symbols, from="2003-01-01", to="2008-01-01")
plot.data<-Cl(GE)
options(repr.plot.width=12, repr.plot.height=8)
chartSeries(plot.data,theme=chartTheme('white'),
    TA=c("addLines(h=31,col='blue',)"))
rm(GE)
```

#### Intermediate Trend

The Khols example is an intermediate down trend from a peak on April in 2015 to October 2015.

```r
symbols=c('KSS')
getSymbols(symbols, from="2015-04-01", to="2015-11-01")
plot.data<-Cl(KSS)
options(repr.plot.width=12, repr.plot.height=8)
chartSeries(plot.data,theme=chartTheme('white'),
    TA="addTA(SMA(plot.data,40), on=1, col='blue')")
rm(KSS)
```

#### Near term trends

Near term trends are three week or less trends. The Swab Growth ETF uptrend is an excellent example of a trend that lasts eight days (just over a week).

```r
symbols=c('SCHG')
getSymbols(symbols, from="2009-10-01", to="2011-10-20")
plot.data<-Cl(SCHG)
options(repr.plot.width=12, repr.plot.height=8)
chartSeries(plot.data["2010-02-01/2010-03-01"],theme=chartTheme('white'),
    TA="addTA(SMA(plot.data,3),col='blue', on=1)")
rm(SCHG)
```

### Support and Resistance

Trends move until the either hit a resistance or support level. Below we have an upward trend that hits resitance levels at 31 and 30.8, and support levels at a little under 30.6 and 30.4.

```r
symbols <- c('GE')
getSymbols(symbols, from="2004-01-01", to="2004-01-15")
plot.data <- Cl(GE)
par(mfrow=c(1,2))
options(repr.plot.width=18, repr.plot.height=6)
chartSeries(plot.data,theme=chartTheme('white'), layout=NULL,
    TA=c("addLines(h=c(31,30.86),col=2)"))
chartSeries(plot.data,theme=chartTheme('white'), layout=NULL,
    TA=c("addLines(h=c(30.4,30.58),col=4,on=1)"))
rm(GE)
```

##### Defining trends

In an uptrend, resistance levels represent pauses in the trend. They are usually exceeded at some point. In similar, but opposite pattern, support levels represent areas that provide temporary pauses in the downward trend.

Support and resistance levels are important to trend analysis. To define an up trend, each successive support level needs to be higher than the previous. Below is an example of successive support levels of an uptrend for the Scwab Growth ETF, trading symbol 'SCHG'. The three horizontal lines show successive support levels that confirm an upward trend.

```r
symbols=c('SCHG')
getSymbols(symbols, from="2009-10-01", to="2011-10-20")
plot.data<-Cl(SCHG)
options(repr.plot.width=12, repr.plot.height=8)
chartSeries(plot.data["2010-02-01/2010-02-27"],theme=chartTheme('white'),
    TA="addLines(h=c(24.6, 24.81, 25.168), on=1, col=2)")
rm(SCHG)
```

##### Defining a trend reversal

When a trend fails to break through a support or resistance level, a trend reversal may be taking place. Below is a chart of Schwab Growth ETF and it shows the tail end of an upward trend. On June 14th the trend hit a resistance level at a little over 78. On the next up tick the trend fails to break through that resistance level, falling short as indicated by the second line. This signals the end of the upward trend and the begining of a down trend.

```r
par(mfrow=c(1,2))
symbols=c('SCHG')
getSymbols(symbols, from="2018-06-01", to="2018-06-30")
plot.data<-Cl(SCHG)
options(repr.plot.width=18, repr.plot.height=6)
chartSeries(plot.data,theme=chartTheme('white'), layout=NULL,
    TA="addLines(h=78.04, on=1, col=4)")
chartSeries(plot.data,theme=chartTheme('white'), layout=NULL,
    TA="addLines(h=c(77.85), on=1, col=2)")
rm(SCHG)
```

### When support and resistance levels reverse roles

Whenever a support or resistance level is penetrated by a significant amount, they reverse their roles and become the opposite. The reasons for this lie in the psychology of the support and resistance levels of the market. In this example, a resistance level was established at 78 on July 11th. Then on July 30th, after trending upwards for a few days, that resistance level was tested (77), only to continue the same trend upwards.

```r
symbols=c('SCHG')
getSymbols(symbols, from="2018-06-01", to="2018-08-30")
plot.data<-Cl(SCHG)
options(repr.plot.width=12, repr.plot.height=8)
chartSeries(plot.data,theme=chartTheme('white'),
    TA="addLines(h=c(78,77.2), on=1, col=2)")
rm(SCHG)
```
